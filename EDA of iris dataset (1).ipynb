
                                                Exploratory Data Analysis on Iris Data Set


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#loading the dataset
df = pd.read_csv('iris_dataset.csv')
```

```python
#display the first few rows of the dataset
print("First few rows of dataset:")
print(df.head())
```

```python
#display basic information anout the dataset
print("\nBasic information about the dataset:")
print(df.info())
```

```python
#summary statistics
print("\nSummary statistics:")
print(df.describe())
```

```python
#check missing values
print("\nCheck for missing values:")
print(df.isnull().sum())
```

```python
#Data Visualization
#Histograms
print("\nHistograms:")
df.hist(figsize=(10,8))
plt.show()
```

```python
#Box plots
print("\nBox plots:")
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.show()
```

```python
#Pair plot
print("\nPair plot:")
sns.pairplot(df)
plt.show()
```

```python
#Heatmap
print("\nHeatmap:")
plt.figure(figsize=(10, 6))
numerical_df = df.select_dtypes(include=['number']) 
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#Identifying patterns
#Group by species and calculate mean of each feature
print("\nMean value by target:")
print(df.groupby("target").mean(numeric_only=True))
```

```python
# Rename the target column to species
df.rename(columns={'target': 'species'}, inplace=True)
# Create the boxplot of species distribution
print("\nSpecies Distribution:")
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
#Sepal vs Petal length
print("\nSepal vs Petal length")
plt.figure(figsize=(10,6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)',hue='species',data=df)
plt.title('Sepal vs Petal length')
plt.show()
```
